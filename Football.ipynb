```python
from tkinter import *
from tkinter.ttk import Combobox
from tkinter import messagebox
from tkinter import *
from PIL import ImageTk,Image
import random
class Football:
    def __init__(self):
        self.root = Tk()
        self.root.state('zoomed')
        self.root.geometry("1000x800")
        self.root.resizable(False, False)

        #frame = Frame(self.root)
        self.canvas = Canvas(self.root,width = 6000, height = 6000)
        self.canvas.pack(expand = YES, fill = BOTH)
        self.file=Image.open("hi1.jpg")
        self.root.title("FOOTBALL")
        image = ImageTk.PhotoImage(self.file)
        self.canvas.create_image(1, 1, image = image, anchor = NW)
        self.do()
        self.root.mainloop()
    def do(self):
        d={'ENG':['Arsenal','Man. City','Man. United'],
        'KAZ':['Astana'],
        'ESP':['Atletico','Real Madrid','Sevilla','Velencia'],
        'BLR':['BATE'],
        'GER':['Leverkusen','Monchenglabach','Monchengladbach'],
        'POR':['Porto'],
        'RUS':['CSKA Moskva'],
        'CRO':['Dinamo Zagreb'],
        'UKR':['Dynamo Kyiv','Shakhtar Donetsk'],
        'TUR':['Galatasaray'],
        'BEL':['Gent'],
        'ITA':['Roma'],
        'FRA':['Lyon'],
        'ISR':['M. Tel-Aviv'],
        'SWE':['Malmo'],
        'GRE':['Olympiacos']}
        country=['ENG','KAZ','ESP','BLR','GER','POR','RUS',
                 'CRO','UKR','TUR','BEL','ITA','FRA','ISR','SWE','GRE']
        D_L_C={'Barcelona':'ESP','Bayern':'GER',
               'Benfica':'POR','Chelsea':'ENG',
               'Juventus':'ITA','Paris':'FRA','PSV':'NED','Zenit':'RUS'}

        dlc=['Barcelona','Bayern','Benfica','Chelsea','Juventus','Paris','PSV','Zenit']
        team=['Arsenal','Man. City','Man. United','Astana','Atletico',
              'Real Madrid','Sevilla','Velencia','BATE','Leverkusen',
              'Monchenglabach','Monchengladbach','Porto','CSKA Moskva',
              'Dinamo Zagreb','Dynamo Kyiv','Shakhtar Donetsk','Galatasaray',
              'Gent','Roma','Lyon','M. Tel-Aviv','Malmo','Olympiacos']
        l=[]
        for i in range(0,8):
            a=[]
            check={'ENG':0,'KAZ':0,'ESP':0,'BLR':0,'GER':0,'POR':0,'RUS':0,'CRO':0,
               'UKR':0,'TUR':0,'BEL':0,'ITA':0,'FRA':0,'ISR':0,'SWE':0,'GRE':0}
            for j in range(0,4):
                if(j==0):
                    v=random.choices(dlc)
                    check[D_L_C[v[0]]]=1  
                    dlc.remove(v[0])
                    a.append(v[0])
                else:
                    while True:
                        v=random.choices(team)
                        for k in country:
                            if v[0] in d[k]:
                                break
                        if(check[k]!=1):
                            check[k]=1
                            break
                    team.remove(v[0])
                    a.append(v[0])
            l.append(a)
            #print(check,"\n")
        ch='A'
        m=0
        self.l2=Label(self.root,text="FOOTBALL TEAM",height=2,width=15,font = "Times 20 bold",bg="light green")
        self.l2.place(x=750,y=10)
        for i in range(0,8):
            b=ord(ch)+i
            c=chr(b)
            a=100
            
            if(i==4):
                m=0
            b=100
            if(i>3):
                b=400
            self.l1=Label(self.root,text="Group"+" "+c+" "+":"+" ",height=1,width=15,font = "Times 12 bold",bg = "light green")
            self.l1.place(x=a+(m+1)*200+200,y=b+50)
            k=0
            for j in range(0,len(l[i])):
                if(j==0):
                    self.l2=Label(self.canvas,text=l[i][j],height=2,width=15,font = "Times 12 bold",fg="red")
                    self.l2.place(x=a+(m+1)*200+200,y=b+30*(k+1)+50)
                else:
                    self.l2=Label(self.canvas,text=l[i][j],height=2,width=15,font = "Times 12 bold")
                    self.l2.place(x=a+(m+1)*200+200,y=b+30*(k+1)+50)
                k+=1
            m+=1
          
b=Football()
```
